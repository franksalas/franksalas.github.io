```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
def get_soup(URL):
    '''input url, returns: soup object of the url '''
    page = requests.get(URL)
    soup = BeautifulSoup(page.text,'html.parser')
    return soup
```

```python
def get_last(sop):
    '''funtion to get the last link for the 
    given bs object to be used to to find the next link and on..
    '''
    l = []
    name_list = sop.find(class_='mw-content-ltr')
    li = name_list.find_all('a')
    for i in li:
        names = i.contents[0]
        links = 'http://starwars.wikia.com' + i.get('href')
        l.append(links)
    #print(type(l[-1]))
    return l[-1]
```

```python
def get_li(sop,d):
    l = []
    name_list = sop.find(class_='mw-content-ltr')
    li = name_list.find_all('a')
    for i in li:
        names = i.contents[0]
        links = 'http://starwars.wikia.com' + i.get('href')
        d.update({names:links})
```
